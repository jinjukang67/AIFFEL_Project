### 로그 변환은 어떻게 분포의 치우침을 줄어들게 만드는가?

```python
# 로그 함수 그래프
xx = np.linspace(0, 10, 500)
yy = np.log(xx)

plt.hlines(0, 0, 10)
plt.vlines(0, -5, 5)
plt.plot(xx, yy, c='r')
plt.show()
```

- 로그함수는 지수함수와 역함수 관계 가짐
- x가 (0,1) 구간에서 기울기가 매우 가파르며 짧다. 
- 반면, y의 구간은 (−∞,0)으로 매우 큼. 
- x가 입력되면 y값은 매우 큰 범위로 벌어진다.
- 로그함수는 입력받은 x값을 넓은 범위로 펼치기 가능
- But, x값이 커짐에 따라 로그 함수의 기울기는 급격히 작아짐. 
- 큰 x값과 y값의 크기 차이가 별로 안나게됨
- 넓은 범위를 가지는 x를 y값의 구간에 모이게함

```python
# 집 가격 데이터 분포
sns.kdeplot(y)
plt.show()
```

- 위 분포는 0 치우쳐져 있는 분포를 보임
- 즉, 0과 1000000 사이에 대부분의 값들이 몰려있고, 아주 소수의 집들이 굉장히 높은 가격을 보인다.
- 따라서 이 분포에 로그 변환을 취하면, 0에 가깝게 몰려있는 데이터들은 넓게 퍼질 것이고, 매우 크게 퍼져있는 소수의 데이터들은 작은 y값으로 모일 것이다.
- 즉, 왼쪽으로 치우친 값들은 보다 넓은 범위로 고르게 퍼지고 오른쪽으로 얇고 넓게 퍼진 값들은 보다 작은 범위로 모이게 되므로 전체 분포는 정규분포의 형상을 띄는 방향으로 변환될 것이다.

```python
# 집 가격 로그 변환 후 확인
y_log_transformation = np.log1p(y)

sns.kdeplot(y_log_transformation)
plt.show() # 정규 분포 모양에 가까워진 모습 확인 가능
```

### RMSE 계산

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


def rmse(y_test,y_pred):
    return np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
# 모델 가져오기
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor

# random_state는 모델초기화나 데이터셋 구성에 사용되는 랜덤 시드값
random_state=2020

gboost = GradientBoostingRegressor(random_state=random_state)
xgboost = XGBRegressor(random_state=random_state)
lightgbm = LGBMRegressor(random_state=random_state)
rdforest = RandomForestRegressor(random_state=random_state)


models = [gboost, xgboost, lightgbm, rdforest]
```

```python
# 클래스 __name__ 속성으로 각 모델 이름 접근 가능
print(gboost.__class__.__name__)
print(xgboost.__class__.__name__)
print(lightgbm.__class__.__name__)
print(rdforest.__class__.__name__)
```

```python
def get_scores(models,train,y):
    
    df = {}

    for model in models:
        # 모델 이름 가져오기
        model_name = model.__class__.__name__
    
        # train, test 데이터셋 분리 - random state 고정
        X_train, X_test, y_train, y_test = train_test_split(train,y,random_state=random_state,test_size=0.2)
    
        # 모델 학습
        model.fit(X_train,y_train)

        # 예측
        y_pred = model.predict(X_test)
    
        # 예측 결과의 rmse값 저장
        df[model_name] = rmse(y_test,y_pred)
    
        # data frame에 저장
        score_df = pd.DataFrame(df,index=['RMSE']).T.sort_values('RMSE',ascending=False)
        
    return score_df

get_scores(models,train,y)
```

### 그리드 탐색

```python
from sklearn.model_selection import GridSearchCV

def my_GridSearch(model, train, y, param_grid, verbose=2, n_jobs=5):

    # GridSearchCV 모델로 초기화
    grid_model = GridSearchCV(model,param_grid=param_grid,\
                         scoring='neg_mean_squared_error',\
                         cv=5,verbose=verbose,n_jobs=n_jobs)
                        #cv = cross validation 하는 횟수
                        # cv 5번 진행하는 이유?
                        # 일반화 오차 추정에 신뢰도 더 높다
    # 학습            
    grid_model.fit(train,y) 
 
    # params, scores 조합 결과값 저장
    params = grid_model.cv_results_['params']
    score = grid_model.cv_results_['mean_test_score']
    
    # params, score로 데이터 프레임 만들기
    param_df = pd.DataFrame(params)
    param_df['score']=score
    
    # -MSE 인 score로 RMSE 컬럼 생성
    param_df['RMSE']=np.sqrt(-1*param_df['score'])
    
    # 로그변환 후 지수변환 안했으므로 RMSE -> RMSLE 이름 변경
   # param_df = param_df.rename(columns={'RMSE':'RMSLE'})
    
    # RMSE 낮은 순서 정렬
    param_df = param_df.sort_values(by='RMSE',ascending=True)
    
    return param_df



# 위에서 만든 함수로 그리드 탐색 해보기
param_grid = {
    'n_estimators':[50,100],
    'max_depth':[1,10],
    
}


model = LGBMRegressor(random_state=random_state)
my_GridSearch(model,train,y,param_grid,verbose=2,n_jobs=5)
```

```python
# 4개 모델로 학습시킨 뒤, AveragingBelnding 예측 결과값 모아 평균한 값 반환
def AveragingBlending(models, x, y, sub_x):
    for m in models : 
        m['model'].fit(x.values, y)
                    # column 옆으로 붙이기
    predictions = np.column_stack([
        m['model'].predict(sub_x.values) for m in models
    ])
    return np.mean(predictions, axis=1)
```
